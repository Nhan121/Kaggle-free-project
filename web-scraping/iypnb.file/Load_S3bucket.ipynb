```python
import boto3, csv, json
import pandas as pd
```

```python
root = "/home/ec2-user/SageMaker"
f = open(f"{root}/notebook_analysis/polar_etl/s3_env_info.json")
data = json.load(f)

bucket_name_to_download = data['s3_bucket_name_input']
key_name_to_dwload = data['key_name_to_dwload']

path = f"s3://{bucket_name_to_download}/{key_name_to_dwload}"
```

### Load data from S3 bucket

```python
s3 = boto3.resource('s3')
s3obj = s3.Object(bucket_name_to_download, key_name_to_dwload).get()
byte_data = s3obj['Body'].read()
str_data = byte_data.decode('utf-8')

csv_data = [row for row in csv.reader(str_data.splitlines())]
header = csv_data[0]
csvdata = csv_data[1:]
pd.DataFrame(data=csvdata, columns=header).head()
```

#### Directly

```python
pd.read_csv(path, header=0, encoding = 'utf-8').head()
```

### Push file to S3

```python
upload_to_df = pd.read_csv(f"{root}/notebook_analysis/polar_etl/upload_draft.csv")
upload_to_df.head()
```

```python
s3 = boto3.resource('s3')
bucket_name_to_upload = data['s3_bucket_name_output']
key_name_to_upload = data['key_name_to_upload']
s3obj = s3.Object(bucket_name_to_upload, key_name_to_upload)
try:
    s3obj.put(Body = upload_to_df.to_csv( index = None, encoding = 'utf-8' ))
    print("Successfully loaded")
except:
    print("Failed to load to S3")
```
